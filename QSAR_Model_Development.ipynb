Molecular Descriptor Calculation

```python
import pandas as pd
from padelpy import from_smiles

# List of SMILES identifiers
smiles_list = [
    "C1=CC=CC=C1",  # Benzene
    "CC(=O)OC1=CC=CC=C1C(=O)O",  # Aspirin
    "C1CCCCC1",  # Cyclohexane
    "C1=CC=C(C=C1)C(=O)O",  # Benzoic Acid
]

# Create an empty DataFrame to store all the descriptors
full_df = pd.DataFrame()

# Loop through each SMILES ID and calculate descriptors with error handling
for index, smile_id in enumerate(smiles_list):
    print(f"Processing SMILES ID {index + 1}/{len(smiles_list)}: {smile_id}")
    
    output_csv = f'{index}.csv'  # Output filename for individual molecule descriptors
    
    try:
        # Calculate descriptors and save to CSV file
        from_smiles(smile_id, fingerprints=True, output_csv=output_csv, timeout=600)
        
        # Read the generated CSV file into a DataFrame and concatenate it to full_df
        df = pd.read_csv(output_csv)
        full_df = pd.concat([full_df, df], ignore_index=True)
    
    except RuntimeError as e:
        print(f"Error processing {smile_id}: {e}")

# Save all descriptors to a single CSV file if any data was collected
if not full_df.empty:
    full_df.to_csv('all_descriptors.csv', index=False)
    print("All descriptors saved to 'all_descriptors.csv'.")
else:
    print("No descriptors were generated.")
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import GaussianNoise, BatchNormalization, Dropout
import pandas as pd
import numpy as np

# Assuming you have 'merge_data1' as your input data, adjust the values as needed
# Preprocess the data to handle null values
X_standardized = X_standardized.fillna(0)  # Replace NaN values with zeros or other suitable values

input_data = keras.Input(shape=(2022,))
noisy_input = GaussianNoise(0.25)(input_data)  # Adjust the standard deviation for more substantial noise
encoded = layers.Dense(500, activation='relu')(noisy_input)

# Adding a Dropout layer for regularization
encoded = Dropout(0.2)(encoded)

decoded = layers.Dense(2022, activation='relu')(encoded)

# This model maps an input to its reconstruction
autoencoder = keras.Model(input_data, decoded)
autoencoder.summary()

optimizer = keras.optimizers.RMSprop(learning_rate=0.001)
autoencoder.compile(optimizer=optimizer, loss='mean_squared_error')

autoencoder.fit(X_standardized, X_standardized,
                epochs=50,
                batch_size=25,
                shuffle=True,
                validation_split=0.3,
                callbacks=[EarlyStopping('val_loss', patience=10)])

# Define the encoder model (up to the bottleneck layer)
encoder = keras.Model(input_data, encoded)

# Define the decoder model
bottleneck_input = keras.Input(shape=(500,))
decoder_layers = autoencoder.layers[-1]
decoder_output = bottleneck_input

# Adding BatchNormalization and Dropout in the decoder
decoder_output = BatchNormalization()(decoder_output)
decoder_output = Dropout(0.2)(decoder_output)

decoder_output = decoder_layers(decoder_output)
decoder = keras.Model(bottleneck_input, decoder_output)

# Get the encoded data
encoded_data = encoder.predict(X_standardized)
dim_reduced_df = pd.DataFrame(encoded_data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping



# Define the neural network model
model = Sequential()
model.add(Dense(500, input_shape=(X_train.shape[1],), activation='relu'))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(Dense(250, activation='relu'))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, Y_train, epochs=100, batch_size=64, validation_split=0.2, verbose=1, 
                    callbacks=[EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True)])

# Evaluate the model on the test set
mse = model.evaluate(X_test, Y_test)
print(f'Mean Squared Error on Test Data: {mse}')

# Make predictions on the test data
Y_pred = model.predict(X_test)

r_squared = r2_score(Y_test, Y_pred)
print(f'R-squared (R²): {r_squared}')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping



# Define the neural network model
model = Sequential()
model.add(Dense(500, input_shape=(X_train.shape[1],), activation='relu'))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(Dense(250, activation='relu'))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='linear'))

# Compile the model with additional hyperparameters
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, Y_train, epochs=100, batch_size=64, validation_split=0.2, verbose=1, 
                    callbacks=[EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True)])

# Evaluate the model on the test set
mse = model.evaluate(X_test, Y_test)
print(f'Mean Squared Error on Test Data: {mse}')

# Make predictions on the test data
Y_pred = model.predict(X_test)

# Calculate metrics
r_squared = r2_score(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
mape = mean_absolute_percentage_error(Y_test, Y_pred)

print(f'R-squared (R²): {r_squared}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Percentage Error (MAPE): {mape}')

# Plot the regression line
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, Y_pred, color='blue', label='Predictions')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', linestyle='--', label='Ideal fit')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Regression Plot')
plt.legend()
plt.show()

# Plot residuals
residuals = Y_test - Y_pred
plt.figure(figsize=(8, 6))
plt.scatter(Y_pred, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()

# Plot training and validation loss curves
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Curves')
plt.legend()
plt.show()
```
