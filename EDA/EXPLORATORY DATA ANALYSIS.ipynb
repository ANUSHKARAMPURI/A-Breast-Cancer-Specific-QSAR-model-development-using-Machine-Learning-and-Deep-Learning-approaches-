```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (assuming it's in a DataFrame called 'df')
df = pd.read_csv(r'D:\GDSC_combinations\breast_anchor_combo.csv\breast_anchor_combo.csv')
```

Distribution of Combo IC50 values for each cell line in the Breast cancer dataset were evaluated and the istogram plots representing the frequency of the distribution were saved succesfully

```python
# List of unique cell lines in your dataset
unique_cell_lines = df['Cell_Line_name'].unique()

# Set custom color palette for Seaborn (you can define your own colors)
custom_palette = [ "#6B6BFF", "#FFD700", "#FF69B4", "#7B68EE", "#32CD32"]

# Set custom style parameters for Seaborn
sns.set_style("whitegrid")
sns.set_palette(custom_palette)
sns.set_context("notebook", font_scale=1.2)  # Adjust the font scale as needed

# Define the directory where you want to save the plots
save_directory = r'D:\GDSC_combinations\breast_anchor_combo.csv\EDA'

# Loop through each cell line and create distribution plots
for cell_line in unique_cell_lines:
    # Subset the data for the current cell line
    subset_data = df[df['Cell_Line_name'] == cell_line]
    
    # Create a histogram or KDE plot (choose one)
    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    sns.histplot(subset_data['Combo IC50'], kde=True)
    
    # Add labels and title
    plt.xlabel('Combo IC50')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.title(f'Distribution of Combo IC50 for {cell_line}')
    
    # Save the plot with a unique name for each cell line
    plot_filename = f'{cell_line}_Combo_IC50_distribution.png'
    plt.savefig(save_directory + plot_filename, dpi=300)  # Adjust dpi for high quality
    
    # Close the current plot to free up memory
    plt.close()
print("Plots saved successfully.")

    
    
```

Among the cel lines and their anchor and library drugs to identify which target pathways were effected more in the existing breast cancer cell lines using Anchor drug and Using Library drug were identified using Heatmaps

HEATMAP FOR ANCHOR DRUGS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set custom color palette for Seaborn (you can define your own colors)
custom_palette = ["#FF6B6B", "#6BFF6B", "#6B6BFF", "#FFD700", "#FF69B4", "#7B68EE", "#32CD32"]

# Set custom style parameters for Seaborn
sns.set_style("whitegrid")
sns.set_palette(custom_palette)
sns.set_context("notebook", font_scale=1.2)  # Adjust the font scale as needed

# List of unique cell lines in your dataset
unique_cell_lines = df['Cell_Line_name'].unique()

# Define the directory where you want to save the plots
save_directory = r'D:\GDSC_combinations\breast_anchor_combo.csv\EDA\TARGET PATHWAYS/'

# Create an empty DataFrame to store summary data
summary_data = pd.DataFrame()

# Loop through each cell line and calculate pathway distribution
for cell_line in unique_cell_lines:
    # Subset the data for the current cell line
    subset_data = df[df['Cell_Line_name'] == cell_line]
    
    # Calculate the pathway distribution for anchor drugs
    anchor_pathway_counts = subset_data['Anchor Pathway'].value_counts().reset_index()
    anchor_pathway_counts.columns = ['Pathway', 'Count']
    
    # Add a column for the cell line
    anchor_pathway_counts['Cell Line'] = cell_line
    
    # Append the pathway counts to the summary data DataFrame
    summary_data = pd.concat([summary_data, anchor_pathway_counts], ignore_index=True)

# Pivot the data to create a heatmap
heatmap_data = summary_data.pivot(index='Cell Line', columns='Pathway', values='Count').fillna(0)

# Create a heatmap
plt.figure(figsize=(15, 20), facecolor='lightgray')
sns.heatmap(heatmap_data, cmap="viridis", annot=False, fmt='g', linewidths=.10, cbar=True)
plt.xlabel('Anchor drug Target Pathway', fontsize=24)
plt.ylabel('Cell Line', fontsize=24)
plt.title('Distribution of Target Pathways for Each Cell Line (Heatmap)', fontsize=16)

# Adjust subplot layout
plt.tight_layout()
plt.show()

# Save the summary heatmap
plot_filename = 'summary_pathway_heatmap.png'
plt.savefig(save_directory + plot_filename, dpi=300)  # Adjust dpi for high quality

print("Summary heatmap saved successfully.")
```

CORRELATION AMONG THE CONTINEOUS VARIABLES PLAYS A KEY ROLE IN DECIDING THE BIAS OF THE VALUES IN ML PREDICTION.

Corelation values of each attribute with its corresponding attribute were identified using heatmap and saved succesfully.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set custom color palette for Seaborn (you can define your own colors)
custom_palette = ["#FF6B6B", "#6BFF6B", "#6B6BFF", "#FFD700", "#FF69B4", "#7B68EE", "#32CD32"]

# Set custom style parameters for Seaborn
sns.set_style("whitegrid")
sns.set_palette(custom_palette)
sns.set_context("notebook", font_scale=1.2)  # Adjust the font scale as needed


# Select the relevant numerical columns starting from the 10th column
numerical_columns = df.columns[11:]

# Calculate the correlation matrix for the selected numerical columns
correlation_matrix = df[numerical_columns].corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
