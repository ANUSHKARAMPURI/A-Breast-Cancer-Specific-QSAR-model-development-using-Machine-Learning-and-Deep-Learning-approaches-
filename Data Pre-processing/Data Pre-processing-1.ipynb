NUMBER OF COMBINATIONS MADE BY THE DRUGS IN BREAST CANCER DATASET WERE ANALYSED BEFORE PERFORMING QSAR MODELLING

```python
import pandas as pd
import itertools
DF = pd.read_excel(r'D:\GDSC_combinations\breast_anchor_combo.csv\breast_cancer_combination_data.xlsx')
```

```python
DF
```

```python
# Create a set to store unique combinations
unique_combinations = set()

# Iterate through the DataFrame
for index, row in DF.iterrows():
    anchor = row["Anchor Name"]
    library = row["Library Name"]
    
    # Check if the reverse combination exists in the set
    reverse_combination = (library, anchor)
    
    # If the reverse combination is not in the set, add the current combination
    if reverse_combination not in unique_combinations:
        unique_combinations.add((anchor, library))

# Print the unique combinations
for combination in unique_combinations:
    print(combination)
```
