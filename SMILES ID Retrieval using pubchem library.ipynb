If we have drug names as a only source for calculating molecular descriptors we can retreive SMILES ID's using pubchem library in python.
before doing so we need to import pbchem library in python
As i am using juyter notebook, i choosed jupyter notebook to install pubchem

```python
pip install pubchem
```

drug names can be gived as a list and the list is used to iterate over each drug molecule and retreive SMILES id from pubchem library

```python
import pubchempy as pcp

compound_names = [
   'AZD7762', 'Pictilisib', '5-Fluorouracil', 'Sapitinib', 'Axitinib',
       'MK-2206', 'PD173074', 'Gemcitabine', 'Nutlin-3a (-)',
       'Temozolomide', 'Luminespib', 'Bortezomib', 'Crizotinib',
       'Cisplatin', 'Trametinib', 'Vorinostat', 'Navitoclax', 'Lapatinib',
       'Taselisib', 'OSI-027', 'MK-1775', 'Palbociclib', 'Linsitinib',
       'Olaparib', 'Docetaxel'
]

smiles_ids = {}

for name in compound_names:
    compounds = pcp.get_compounds(name, 'name')
    if compounds:
        smiles_ids[name] = compounds[0].isomeric_smiles

for name, smiles in smiles_ids.items():
    print(f"{name}: {smiles}")

```
